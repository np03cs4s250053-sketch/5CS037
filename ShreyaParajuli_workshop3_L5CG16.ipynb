```python
#3.1
import pandas as pd
import os
import requests

# Define the path and URL for the titanic.csv file
file_path = "/content/sample_data/titanic.csv"
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Check if the file already exists, if not, download it
if not os.path.exists(file_path):
    print(f"Downloading {os.path.basename(file_path)}...")
    response = requests.get(url)
    response.raise_for_status() # Raise an error for bad status codes
    with open(file_path, 'wb') as f:
        f.write(response.content)
    print(f"Successfully downloaded {os.path.basename(file_path)}.")
else:
    print(f"{os.path.basename(file_path)} already exists.")

titanic = pd.read_csv(file_path)
print(titanic.info())
print(titanic.head())

# 1.
fare = titanic[['Fare']]
print("Fare column:\n", fare.head())

class_age = titanic[['Pclass', 'Age']]
print("Pclass and Age columns:\n", class_age.head())


survived_gender = titanic[['Survived', 'Sex']]
print("Survived and Sex columns:\n", survived_gender.head())
```

```python
# Problem 2 - Subsetting

fare_gt_100 = titanic[titanic['Fare'] > 100]
print("Fare > 100:\n", fare_gt_100)
```

```python

first_class = titanic[titanic['Pclass'] == 1]
print("First class passengers:\n", first_class)

```

```python
female_under_18 = titanic[(titanic['Age'] < 18) & (titanic['Sex'] == 'female')]
print("Female passengers under 18:\n", female_under_18)

```

```python
embarked_c_or_s = titanic[titanic['Embarked'].isin(['C', 'S'])]
print("Embarked at C or S:\n", embarked_c_or_s)

```

```python
first_second_class = titanic[titanic['Pclass'].isin([1, 2])]
print("First and Second class passengers:\n", first_second_class)
```

```python
#3.2
median_age = titanic['Age'].median()
titanic['Age'] = titanic['Age'].fillna(median_age)

titanic['fare_per_year'] = titanic['Fare'] / titanic['Age']

high_fare_age = titanic[titanic['fare_per_year'] > 5]

high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

result = high_fare_age_srt[['Name', 'fare_per_year']]
print("Passengers with highest fare relative to age:\n", result.head())

titanic['fare_per_class'] = titanic['Fare'] / titanic['Pclass']

adult_males = titanic[(titanic['Sex'] == 'male') & (titanic['Age'] >= 18)]


adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)


result_adult_male = adult_males_srt[['Name', 'Age', 'fare_per_class']]
print("Adult male passengers with highest fare relative to class:\n", result_adult_male.head())
```

```python
#3.3
#problem1
total_fare = titanic['Fare'].sum()

fare_by_class = titanic.groupby('Pclass')['Fare'].sum()
fare_percentage_by_class = (fare_by_class / total_fare) * 100
print("Percent of total fare revenue by class:\n", fare_percentage_by_class)

#problem2
def age_group(age):
    if age < 18:
        return 'child'
    elif 18 <= age < 65:
        return 'adult'
    else:
        return 'senior'

titanic['age_group'] = titanic['Age'].apply(age_group)

total_passengers = len(titanic)
passengers_by_age_group = titanic['age_group'].value_counts()
age_group_percentage = (passengers_by_age_group / total_passengers) * 100
print("Percent of passengers by age group:\n", age_group_percentage)
```
