```python
import numpy as np
A = np.array([[3, 4],[7, 8]])
B = np.array([[5, 3],[2, 1]])
```

```python
A_inv = np.linalg.inv(A)
proof1 = np.dot(A, A_inv)
print(proof1)
```

```python
AB = np.dot(A, B)
BA = np.dot(B, A)
print(AB)
print(BA)
```

```python
AB_transpose = (AB).T
BT_AT = np.dot(B.T, A.T)
print(AB_transpose)
print(BT_AT)
```

4. Matrix Multiplication
• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.
• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Mea-
sure and print the time taken for this operation.

```python
import numpy as np
import time

#ADDITION
# Python lists
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))
start = time.time()
list_add = [list1[i] + list2[i] for i in range(1_000_000)]
end = time.time()
print("Python List Addition Time:", end - start, "seconds")
# NumPy arrays
arr1 = np.arange(1_000_000)
arr2 = np.arange(1_000_000)
start = time.time()
arr_add = arr1 + arr2
end = time.time()
print("NumPy Array Addition Time:", end - start, "seconds")

#MULTIPLICATION
# Python lists
start = time.time()
list_mul = [list1[i] * list2[i] for i in range(1_000_000)]
end = time.time()
print("Python List Multiplication Time:", end - start, "seconds")
# NumPy arrays
start = time.time()
arr_mul = arr1 * arr2
end = time.time()
print("NumPy Array Multiplication Time:", end - start, "seconds")

#DOT PRODUCT
# Python lists
start = time.time()
dot_py = sum(list1[i] * list2[i] for i in range(1_000_000))
end = time.time()
print("Python List Dot Product Time:", end - start, "seconds")
# NumPy arrays
start = time.time()
dot_np = np.dot(arr1, arr2)
end = time.time()
print("NumPy Dot Product Time:", end - start, "seconds")

#MATRIX MULTIPLICATION
# Python lists
M1 = [[i for i in range(1000)] for _ in range(1000)]
M2 = [[i for i in range(1000)] for _ in range(1000)]
start = time.time()
# matrix multiplication using python
result_py = [[sum(M1[i][k] * M2[k][j] for k in range(1000))
              for j in range(1000)]
              for i in range(1000)]
end = time.time()
print("Python List Matrix Multiplication Time:", end - start, "seconds")

# NumPy arrays
A = np.arange(1_000_000).reshape(1000, 1000)
B = np.arange(1_000_000).reshape(1000, 1000)
start = time.time()
result_np = np.dot(A, B)
end = time.time()
print("NumPy Matrix Multiplication Time:", end - start, "seconds")
```
