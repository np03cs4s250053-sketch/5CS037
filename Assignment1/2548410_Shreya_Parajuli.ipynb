```python
def classify_hdi(hdi):
    if hdi < 0.550:
        return "Low"
    elif hdi < 0.700:
        return "Medium"
    elif hdi < 0.800:
        return "High"
    else:
        return "Very High"
```

```python
# Composite Development Score:
# â€¢ Create a new metric called Composite Score using the formula:
#   Composite Score = 0.30 Ã— Life Expectancy Index + 0.30 Ã— GNI per Capita Index}
# Here: Life Expectancy Index â†’ "life expectancy" and GNI per Capita Index â†’ "gross inc percap"â€¢ Rank South Asian countries based on Composite Score.
# â€¢ Plot the top 5 countries in a horizontal bar chart.
# â€¢ Compare the ranking of countries by Composite Score with their HDI ranking and discuss any
# differences

south_asia_df['Composite Score'] = (
    0.30 * south_asia_df['life_expectancy'] +
    0.30 * south_asia_df['gross_inc_percap']
)
```

```python
# IQR for HDI
Q1_hdi = south_asia_latest['hdi'].quantile(0.25)
Q3_hdi = south_asia_latest['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

lower_hdi = Q1_hdi - 1.5 * IQR_hdi
upper_hdi = Q3_hdi + 1.5 * IQR_hdi

# IQR for GNI per Capita
Q1_gni = south_asia_latest['gross_inc_percap'].quantile(0.25)
Q3_gni = south_asia_latest['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

lower_gni = Q1_gni - 1.5 * IQR_gni
upper_gni = Q3_gni + 1.5 * IQR_gni

# Identify outliers
outliers = south_asia_latest[
    (south_asia_latest['hdi'] < lower_hdi) |
    (south_asia_latest['hdi'] > upper_hdi) |
    (south_asia_latest['gross_inc_percap'] < lower_gni) |
    (south_asia_latest['gross_inc_percap'] > upper_gni)
]

outliers[['country', 'hdi', 'gross_inc_percap']]
```

```python
# 7. Outlier Detection:
# â€¢ Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using the 1.5 Ã— IQR rule.
# â€¢ Create scatter plots highlighting outliers in a different color.
# â€¢ Discuss the significance of these outliers.
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower) | (df[column] > upper)]
    return outliers
```

```python
sa_hdi_outliers = detect_outliers_iqr(south_asia_2020_2022, 'hdi')
sa_gni_outliers = detect_outliers_iqr(south_asia_2020_2022, 'gross_inc_percap')

sa_hdi_outliers[['country', 'year', 'hdi']]
```
